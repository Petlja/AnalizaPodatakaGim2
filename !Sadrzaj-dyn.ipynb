# Рачунарство и информатика за 2. гимназије

**Циљ** предмета Рачунарство и информатика за 2. гимназије је „стицање знања, овладавање вештинама, формирање вредносних ставова који доприносе развоју информатичке писмености неопходне за даље школовање, живот и рад у савременом друштву“.

Курс ставља фокус на

- податке,
- обраду података, и
- визуелизацију података

у три различита контекста:

- програми за табеларна израчунавања,
- рад са подацима у текстуалном програмском језику (Џупитер), и
- програмирање у програмском језику Пајтон.

Овде се налази део курса који се односи на Џупитер и Пајтон у облику Џупитер радних свески које могу да се покрену, што омогућује да се код извршава интерактивно и да се на тај начин стекну одговарајуће вештине.
